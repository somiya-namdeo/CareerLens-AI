# CareerLens AI - Job Role Prediction Module

## Objective

The objective of this module is to predict the most suitable job role for a candidate based on the skills, experience, education, certifications, and resume information provided. This helps recruiters and organizations identify the best-fit role for candidates while assisting job seekers in understanding their career alignment.

## Dataset

Job Role Prediction Dataset containing candidate profiles, skills, educational background, work experience, certifications, and target job roles.

## Machine Learning Models Used

* Logistic Regression
* Random Forest Classifier
* Naive Bayes
* Support Vector Machine (SVM)
* K-Nearest Neighbors (KNN)

## Text Processing Technique

* Text Cleaning
* TF-IDF (Term Frequency - Inverse Document Frequency) Vectorization

## Evaluation Metrics

* Accuracy
* Precision
* Recall
* F1 Score
* Classification Report
* Confusion Matrix

## Output

The best-performing model is saved as:

* job_role_model.pkl
* job_role_tfidf.pkl

## Application in CareerLens AI

This module serves as the core career intelligence engine of CareerLens AI. After analyzing candidate information and resume content, it predicts the most suitable job role and provides career insights, recommendations, and workforce intelligence reports.
